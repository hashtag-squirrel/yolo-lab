```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')

folder = 'combined'
imgsz = 640
epochs = 50
batch = 8
save_folder_name = f'yolov8_{epochs}_{folder}'

results = model.train(
    data=f'data/{folder}/data.yaml',
    imgsz=imgsz,
    epochs=epochs,
    batch=batch,
    name=save_folder_name
)
```
